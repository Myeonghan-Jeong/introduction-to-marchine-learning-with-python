## 2.1 분류와 회귀

- 지도 학습은 크게 **분류**와 **회귀**로 나눕니다.

    - 분류는 미리 정의된 여러 **클래스 레이블** 중 하나를 예측하는 문제입니다.
    
    > 앞서서 살펴본 붓꽃 예측하기가 분류 문제 중 하나입니다.
    
- 분류는 두 개의 클래스로 분류하는 **이진 분류**와 셋 이상 클래스로 분류하는 **다중 분류**로 나뉩니다.

    - 이진 분류는 답이 예 혹은 아니오로 나오도록 만드는 것이라고 생각하면 됩니다.

    > 예를 들어 특정 이메일이 스팸 메일인지 아닌지 판별하는 것이 이진 분류 문제 중 하나입니다.

#### Note

- 이진 분류에서 한 클래스를 **양성** 클래스, 다른 하나를 **음성** 클래스라고 합니다.

    - 양성 클래스라고 해서 좋은 값이나 장점을 말하는 것은 아니며 학습하고자 하는 대상을 의미합니다.
    
    > 예를 들면 스팸 메일의 경우 양성이 스팸 클래스 입니다.

- 두 클래스 중 어떤 것을 양성으로 정할지는 주관적인 문제로 분야에 따라 달라집니다.

- 분류 문제 중 다른 하나는 다중 분류로 여러 개의 정답 중 하나를 찾는 것을 의미합니다.

    - 즉, 다중 분류는 답이 여러 개가 존재하는 것을 의미합니다.

    > 앞서서 본 붓꽃 분류 문제나 해당 글이 어떤 언어로 작성된 글인지 맞추는 문제 등이 다중 분류 문제입니다.

- 회귀는 연속적인 숫자 또는 프로그래밍 용어로 **부동소수점수**(수학 용어로 **실수**)를 예측하는 것을 말합니다.

    - 예를 들어 어떤 사람의 교육 수준, 나이, 주거지를 바탕으로 연간 소득을 예측 하는 것이 회귀 문제 중 하나입니다.
    
    - 그 외에도 옥수수 농장의 전년도 수확량과 날씨, 고용 인원 수 등을 바탕으로 올해 수확량을 예측하는 것도 한 문제입니다.

- 출력 값에 연속성이 있는지 질문해보면 회귀와 분류 문제를 쉽게 구분할 수 있습니다.

    - 예상 출력 값 사이에 연속성이 있다면 회귀 문제입니다.
    
        - 예를 들어 연소득을 예측하는 문제는 예측한 정답에서 1 ~ 2원 차이가 났다고 해서 큰 문제가 일어나지 않습니다.
        
        - 반면에 언어를 맞추는 문제는 어느 정도란 것이 없습니다. 즉, 언어는 어느 한 언어가 아니라면 다른 언어입니다.

## 2.2 일반화, 과대적합, 과소적합

- 지도 학습에서 훈련 데이터로 학습한 모델이 훈련 데이터와 특성이 같다면 처음 보는 새로운 데이터도 정확히 예측할 거라 기대합니다.

- 모델이 처음 보는 데이터에 대해 정확하게 예측할 수 있으면 이를 훈련 세트에서 테스트 세트로 **일반화** 되었다고 합니다.

    - 그래서 모델을 만들 때는 가능한 한 정확하게 일반화되도록 해야 합니다.

- 보통 훈련 세트에 대해 정확히 예측하도록 모델을 구축합니다. 이 때 훈련 세트와 테스트 세트가 매우 비슷하다면 그 모델이 테스트 세트에서도 정확히 예측하리라 기대할 수 있습니다.

- 그러나 항상 그런 것만은 아닙니다. 예로 아주 복잡한 모델을 만들면 훈련 세트에만 정확한 모델이 되어버릴 수 있습니다.

- 알고리즘이 새로운 데이터도 잘 처리하는지 측정하는 방법은 테스트 세트로 평가해보는 것밖에 없습니다.

    - 그러나 직관적으로 보더라도 간단한 모델이 새로운 데이터에 더 일반화될 것이라고 예상할 수 있습니다.

- 이에 비추어 가진 정보를 모두 사용해 너무 복잡한 모델을 만드는 것을 **과대적합**이라고 합니다.

    - 과대 적합은 모델이 훈련 세트의 각 샘플에 너무 가깝게 맞춰져서 새로운 데이터에 일반화되기 어려울 때 일어납니다.

- 반대로 너무 간단한 모델은 데이터의 면면과 다양성을 잡아내지 못할 것이며 훈련 세트와도 맞지 않을 것입니다. 이를 **과소적합**이라고 합니다.

- 즉, 모델이 복잡할수록 훈련 데이터에 대해 더 정확히 예측할 수 있으나 너무 복잡해지면 훈련 세트의 각 데이터 포인트에 너무 민감해져 새로운 데이터에 잘 일반화되지 못합니다.

> 즉, 우리가 만들어야하는 모델은 일반화 성능이 최대가 되는 최적점에 있는 모델입니다.

### 2.2.1 모델 복잡도와 데이터셋 크기의 관계

- 모델 복잡도는 훈련 데이터셋에 담긴 입력 데이터의 다양성과 관련이 깊습니다.

    - 데이터셋에 다양한 데이터 포인트가 많을수록 과대적합 없이 더 복잡한 모델을 만들 수 있습니다.
    
    - 보통 데이터 포인트를 더 많이 모으는 것이 다양성을 키워주므로 큰 데이터셋은 더 복잡한 모델을 만들 수 있게 해줍니다.
    
    - 다만, 같은 데이터 포인트를 중복하거나 매우 비슷한 데이터를 모으는 것은 도움이 되지 않습니다.

- 데이터를 더 많이 수집하고 적절하게 더 복잡한 모델을 만들면 지도 학습 문제에서 놀라운 결과를 얻을 수 있습니다.

    > 이 책에서는 고정 크기 데이터셋을 사용할 것입니다.
    >
    > 이는 모델을 변경하거나 조정하는 것보다 이득일 수 있으나 실제 환경에서는 데이터를 얼마나 많이 모을지 정해야 합니다.

### 2.3.1 예제에 사용할 데이터셋

- 여러 알고리즘을 설명하기 위해 여러 데이터셋을 사용할 예정입니다.

    > 이 데이터셋은 작고 인위적으로 만든 것으로 알고리즘의 특징을 부각하기 위해 만든 것도 있으니 주의하세요.

- 두 개의 특성을 가진 forge 데이터셋은 인위적으로 만든 이진 분류 데이터셋입니다.

    - 다음 코드로 해당 데이터셋의 모든 데이터 포인트를 산점도로 그려보세요. x축은 첫 번째 특성, y축은 두 번째 특성입니다.
    
    - 산점도 그래프의 점 하나가 각 데이터 포인트이며 점의 색과 모양은 데이터 포인트가 속한 클래스입니다.

```python
# 이 행을 실행해서 필요한 라이브러리들을 먼저 import 하세요!
from IPython.display import display
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import mglearn
```

```python
# 데이터셋을 만듭니다.
X, y = mglearn.datasets.make_forge()

# 산점도를 그립니다.
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(['class 0', 'class 1'], loc=4)
plt.xlabel('first character')
plt.ylabel('second character')
print('X.shape:', X.shape)
```

- X.shape 값에서 알 수 있듯이 이 데이터셋은 26개의 데이터 포인트와 2개의 특성을 가지고 있습니다.

    - wave 데이터셋은 입력 특성 하나와 모델링할 타깃 변수를 가집니다.

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
plt.plot(X, y, 'o')
plt.ylim(-3, 3)
plt.xlabel('chracter')
plt.ylabel('target')
```

- 종이는 2차원이라 둘이 넘는 특성은 표현하기 어려우니 손쉽게 시각화하기 위해서 간단한 저차원 데이터셋을 사용하겠습니다.

- 중요한 점은 특성이 적은 데이터셋(**저차원 데이터셋**)에서 얻은 직관이 특성이 많은 데이터셋(**고차원 데이터셋**)에서 그대로 유지되지 않을 수 있다는 점입니다.

    - 이 점을 숙지하고 알고리즘을 배울 때는 저차원 데이터셋을 사용하는 것이 낫습니다.

### 2.3.2 k-최근접 이웃

- k-NN 알고리즘은 가장 간단한 머신러닝 알고리즘입니다. 훈련 데이터셋을 그냥 저장하는 것이 모델을 만드는 과정의 전부입니다.

- 새로운 데이터 포인트에 대해 예측할 땐 알고리즘이 가장 가까운 데이터 포인트, 즉 '최근접 이웃'을 찾습니다.

#### k-최근접 이웃 분류

- 가장 간단한 k-NN 알고리즘은 가장 가까운 훈련 데이터 포인트 하나를 최근접 이웃으로 찾아 예측에 사용합니다.

- 단순히 이 훈련 데이터 포인트의 출력이 예측되면 다음과 같이 분류할 수 있습니다.

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

- 위 그래프는 데이터 포인트 3개(별 모양)에 대해 각각 가장 가까운 훈련 데이터 포인트 1개를 연결한 것입니다.

    - 이 반식이 1-최근접 이웃 알고리즘의 예측으로 데이터 포인트의 레이블이 됩니다.

- 반면 가장 가까운 이웃을 하나가 아닌 임의의 k개를 선택할 수도 있습니다. 바로 이 경우가 k-최근접 이웃 알고리즘입니다.

    - 둘 이상의 이웃을 선택할 때는 레이블을 정하기 위한 투표를 합니다.
    
    - 즉, 테스트 포인트 하나에 대해 각 클래스에 속한 이웃이 몇 개인지 셉니다. 그리고 이웃이 더 많은 클래스를 레이블로 지정합니다.
    
    - 다시 말해 k-최근접 이웃 중 다수의 클래스가 레이블이 됩니다. 다음은 세 개의 최근접 이웃을 사용하는 예입니다.

```python
mglearn.plots.plot_knn_classification(n_neighbors=3)
```

- 여기서도 예측은 연결된 데이터 포인트의 색으로 나타납니다. 이를 살펴보면 이웃을 하나만 선택할 때와 결과가 달라진 데이터를 알 수 있습니다.

- 이 예시는 이진 분류 문제이지만 클래스가 다수인 데이터셋도 같은 방법을 적용할 수 있습니다.

    - 클래스가 여러 개 일때도 각 클래스에 속한 이웃이 몇 개인지를 헤아려 가장 많은 클래스를 예측값으로 사용합니다.

> 이제 scikit-learn을 사용해 k-최근접 이웃 알고리즘을 어떻게 적용하는지 살펴볼 것입니다.

```python
from sklearn.model_selection import train_test_split

# 일반화 성능을 평가하기 위해 데이터를 훈련 세트와 테스트 세트로 나눕니다.
X, y = mglearn.datasets.make_forge()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# k-NN 클래스를 import하고 하나의 객체로 만듭니다. 이 때 이웃의 수 같은 매개변수를 지정합니다.
from sklearn.neighbors import KNeighborsClassifier

clf = KNeighborsClassifier(n_neighbors=3)

# 훈련 세트를 사용해 분류 모델을 학습시킵니다.
# k-NN의 경우 에측할 때 이웃을 찾을 수 있도록 데이터를 저장합니다.
clf.fit(X_train, y_train)

# 테스트 세트를 predict를 사용해 예측합니다.
# k-NN은 각 데이터 포인트에 대해 훈련 세트에서 가장 가까운 이웃을 계산한 다음 가장 많은 클래스를 찾습니다.
print('테스트 세트 예측:', clf.predict(X_test))

# 모델이 얼마나 잘 일반화되었는지 score를 사용해 평가합니다.
print(f'테스트 세트 정확도: {clf.score(X_test, y_test):.2f}')
```

#### KNeighborsClassifier 분석

- 2차원 데이터셋이므로 가능한 모든 테스트 포인트의 예측을 xy 평면에 그릴 수 있습니다.

    - 각 데이터 포인트가 속한 클래스에 따라 평면에 색을 칠하여 알고리즘이 지정한 클래스로 나뉘는 **결정 경계**를 볼 수 있습니다.
    
> 다음 코드는 이웃이 1, 3, 9개일 때 결정 경계를 보여줍니다.

```python
fig, axes = plt.subplots(1, 3, figsize=(10, 3))

for n_neighbors, ax in zip([1, 3, 9], axes):
    # fit 메서드는 self 오브젝트를 리턴하므로 객체 생성과 fit을 동시에 할 수 있습니다.
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X, y)
    
    # 그래프를 그립니다.
    mglearn.plots.plot_2d_separator(clf, X, fill=True, eps=0.5, ax=ax, alpha=.4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    
    # 각 그래프의 속성을 부여합니다.
    ax.set_title(f'{n_neighbors} neighbors')
    ax.set_xlabel('attribute 0')
    ax.set_ylabel('attribute 1')

axes[0].legend(loc=3)
```

- 위 그래프를 보면 이웃이 적으면 결정 경계가 훈련 데이터에 가깝게 따라가고 많으면 경계가 부드러워짐을 알 수 있습니다.

- 다시 말해 이웃을 적게 사용하면 모델의 복잡도가 높아지고 이웃을 많이 사용하면 복잡도가 낮아집니다.

    - 극단적으로 이웃의 수를 훈련 데이터 전체로 지정하면 모든 테스트 포인트가 같은 이웃을 가져 테스트 포인트에 대한 예측이 같은 값이 됩니다.
    
    - 즉, 훈련 세트에서 가장 많은 데이터 포인트를 가진 클래스가 예측값이 됩니다.

- 이어서 모델의 복잡도와 일반화 사이의 관계를 실제 데이터인 유방암 데이터셋을 사용해 확인하겠습니다.

```python
from sklearn.datasets import load_breast_cancer

# 데이터 객체를 저장합니다.
cancer = load_breast_cancer()

# 데이터셋을 훈련 세트와 테스트 세트로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify=cancer.target, random_state=66)

# 정확도를 평가합니다.
training_accuracy, test_accuracy = [], []
neighbors_setting = range(1, 11)
for n_neighbors in neighbors_setting:
    clf = KNeighborsClassifier(n_neighbors=n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))

# 그래프를 그립니다.
plt.plot(neighbors_setting, training_accuracy, label='training accuracy')
plt.plot(neighbors_setting, test_accuracy, label='test accuracy')
plt.xlabel('n_neighbors')
plt.ylabel('accuracy')
plt.legend()
```

- 위 그래프에 따르면 n_neighbors 수에 따를 훈련 세트와 테스트 세트의 정확도를 알 수 있습니다.

- 실제 그래프는 이렇게 매끈하게 나오지 않지만 여기서 과대적합과 과소적합의 특징을 볼 수 있습니다.

    - 최근접 이웃의 수가 하나일 때 훈련 데이터에 대한 예측이 완벽하나 이웃의 수가 늘어나면 모델은 단순해져 훈련 세트의 정확도가 줄어듭니다.
    
    - 반면에 이웃을 하나만 사용한 테스트 세트의 정확도는 이웃을 많이 사용할 때보다 낮습니다.
    
    - 이는 모델이 너무 복잡함을 설명하며 반대로 이웃이 많을 땐 모델이 너무 단순해 정확도가 나빠집니다.
    
    > 즉, 정확도가 가장 좋을 때는 중간 정도인 6개를 사용한 경우입니다.
