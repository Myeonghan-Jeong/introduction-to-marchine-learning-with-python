## 3.1 비지도 학습의 종류

- 이번 장에서는 두 가지 비지도 학습을 살펴보겠습니다. 바로 데이터의 비지도 변환과 군집입니다.

- **비지도 변환**은 데이터를 새롭게 표현하여 사람이나 머신러닝 알고리즘이 원래 데이터보다 쉽게 해석할 수 있도록 만드는 알고리즘입니다.

    - 비지도 변환이 널리 사용되는 분야는 많은 고차원 데이터의 특성 수를 줄이면서 꼭 필요한 특징만 포함한 데이터로 표현하는 차원 축소로 대표적으로 시각화하기 위해 데이터셋을 2차원으로 변경하는 경우입니다.
    
    - 비지도 변환에서 데이터를 구성하는 단위나 성분을 찾기도 하는데 많은 텍스트 문서에서 주제를 추출하는 것이 이런 예시 입니다. 이 때 처리할 작업은 문서에서 이야기하는 주제들이 무엇인지 찾고 학습하는 것입니다.

- **군집 알고리즘**은 데이터를 비슷한 것끼리 그룹으로 묶는 것으로 대표적으로 소셜 미디어 사이트에서 업로드한 사진을 자동으로 분류하는 것입니다.

## 3.2 비지도 학습의 도전 과제

- 비지도 학습에서 가장 어려운 일은 알고리즘이 뭔가 유용한 것을 학습했는지 평가하는 것입니다.

    - 비지도 학습은 보통 레이블이 없는 데이터에 적용하기 때문에 무엇이 올바른 출력인지 모릅니다. 그래서 어떤 모델이 일을 잘하고 있는지 판단하기가 매우 어렵습니다.
    
    - 더욱이 비지도 학습 모델에 우리가 원하는 것을 알려줄 방법이 없기 때문에 비지도 학습의 결과를 평가하기 위해서는 직접 확인하는 것이 유일한 방법일 때가 많습니다.

- 이런 이유로 비지도 학습 알고리즘은 데이터 과학자가 데이터를 더 잘 이해하고 싶을 때 탐색적 분석 단계에서 많이 사용하며 지도 학습의 전처리 단계에서도 많이 사용합니다.

    - 비지도 학습의 결과로 새롭게 표현된 데이터를 사용해 학습하면 지도 학습의 정확도가 좋아지기도 하고 메모리와 시간을 절약할 수 있습니다.

- 진짜 비지도 학습 알고리즘으로 들어가기 전에 몇 가지 간단한 전처리 메서드를 짧게 설명하겠습니다. 지도 학습 알고리즘에서 전처리와 스케일 조정을 자주 사용하지만, 스케일 조정 메서드는 지도 정보를 사용하지 않으므로 비지도 방식입니다.

## 3.3 데이터 전처리와 스케일 조정

- 앞 장에서 본 신경망과 SVM 같은 알고리즘은 데이터의 스케일에 매우 민감합니다. 그래서 보통 이런 알고리즘에 맞게 데이터의 특성 값을 조정합니다.

    - 보통 특성마다 스케일을 조정해서 데이터를 변경합니다.

```python
import mglearn

mglearn.plots.plot_scaling()
```

### 3.3.1 여러 가지 전처리 방법

- 앞 선 그림의 첫 번째 그래프는 두 개의 특성을 인위적으로 만든 이진 분류 데이터셋으로 x축 10 ~ 15, y축 1 ~ 9 사이에 데이터가 존재합니다.

- 그 다음 오른쪽 네 개의 그래프는 이들을 기준이 되는 범위로 변환하는 방법들로 scikit-learn의 StandardScaler는 각 특성의 평균을 0, 분산을 1로 변경하여 모든 특성이 같은 크기를 가지게 합니다.

- 그러나 이 방법은 특성의 최솟값과 최댓값을 제한하지 않는데 RobustScaler는 특성들이 같은 스케일을 갖게 하는 점에서 StandardScaler와 비슷하지만 중간값과 사분위값을 사용해 이를 처리합니다.
    
    - 이런 방식 덕분에 RobustScaler는 전체 데이터와 아주 동떨어진 데이터 포인트인 **이상차**에 영향을 받지 않습니다.
    
- 반면에 MinMaxScaler는 모든 특성이 정확하게 0과 1사이에 위치하도록 데이터를 변경하며 모든 데이터가 x, y 값이 0과 1사이로 조정됩니다.

- 마지막으로 Normalizer는 매우 다른 스케일 조정 기법으로 특성 벡터의 유클라디안 길이가 1이 되도록 데이터 포인트를 조정합니다.

    - 다시 말해 각 데이터 포인트가 다른 비율로 길이에 반비례해 스케일이 조정된다는 뜻으로 정규화는 특성 벡터의 길이는 상관없고 데이터의 방향 혹은 각도가 중요할 때 많이 사용합니다.

### 3.3.2 데이터 변환 적용하기

- 여러 종류의 변환을 scikit-learn을 사용해 실제로 적용해보겠습니다. 데이터셋은 2장에서 사용한 cancer를 사용합니다.

    - 스케일을 조정하는 전처리 메서드들은 보통 지도 학습 알고리즘을 적용하기 전에 적용하며 여기서는 cancer에 커널 SVM(SVC)를 적용하고 전처리에 MinMaxScaler를 사용하겠습니다.

- 먼저 데이터를 적재하고 훈련 세트와 테스트 세트로 나눕니다.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=1)

print(X_train.shape, X_test.shape)
```

- 이 데이터셋에는 569개의 데이터 포인트가 있고 각 데이터 포인트는 30개의 측정값으로 이뤄져 있습니다. 여기서는 426개를 훈련 세트로, 143개를 테스트 세트로 나누었습니다.

- 이전에 만든 지도 학습 모델처럼 먼저 전처리가 구현된 python 클래스를 import하고 객체를 생성합니다.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
```

- 그런 다음 fit 메서드를 훈련 데이터에 적용하는데 MinMaxScaler의 fit은 훈련 세트의 특성마다 최솟값과 최댓값을 계산합니다. 2장의 분류 모델이나 회귀 모델과는 달리 스케일 객체는 훈련 데이터만 사용하며 결과 레이블은 사용하지 않습니다.

```python
scaler.fit(X_train)
```

- fit 메서드로 학습한 변환을 적용하려면, 즉 실제 데이터의 스케일을 조정하려면 스케일 객체의 transform 메서드를 사용합니다. 이는 새로운 데이터 표현을 만들 때 사용합니다.

```python
# 데이터 변환 과정입니다.
X_train_scaled = scaler.transform(X_train)

# 결과를 확인합니다.
print('변환 후 크기:', X_train_scaled.shape)
print('스케일 변환 전 특성별 최소값:\n', X_train.min(axis=0))
print('스케일 변환 전 특성별 최대값:\n', X_train.max(axis=0))
print('스케일 변환 후 특성별 최소값:\n', X_train_scaled.min(axis=0))
print('스케일 변환 후 특성별 최대값:\n', X_train_scaled.max(axis=0))
```

- 변환된 데이터의 배열 크기는 원래 데이터와 동일하므로 특성 값이 이동되거나 크기가 조정되기만 했음을 보입니다. 예상대로 모든 특성 값은 0과 1 사이가 되었습니다.

- 이 데이터에 SVM을 적용하려면 테스트 세트도 변환해야 합니다. 이 땐 X_test를 transform 메서드에 넣고 활용합니다.

```python
# 데이터 변환 과정입니다.
X_test_scaled = scaler.transform(X_test)

# 결과를 확인합니다.
print('스케일 변환 후 특성별 최소값:\n', X_test_scaled.min(axis=0))
print('스케일 변환 후 특성별 최대값:\n', X_test_scaled.max(axis=0))
```

- 보시다시피 스케일을 조정한 테스트 세트의 최솟값과 최댓값이 0과 1이 아니며 일부 특성은 범위를 벗어났습니다.

- 이는 MinMaxScaler와 다른 모든 스케일 모델도 항상 훈련 세트와 테스트 세트에 같은 변환을 적용하기 때문입니다.

    - transform 메서드는 테스트 세트의 최솟값과 최댓값의 범위가 아닌 훈련 테스트의 최솟값을 뺀 뒤 훈련 세트의 범위로 나눕니다.
